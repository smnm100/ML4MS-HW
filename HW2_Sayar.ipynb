```python
import numpy as np

# function F(x,y) and complex conjugate F*(x,y)
def F(x, y):
    return np.exp(1j*y) + 2 * np.exp(-0.5j*y) * np.cos(0.5*np.sqrt(3) * x)

def F_conj(x, y):
    return np.exp(-1j*y) + 2 * np.exp(0.5j*y) * np.cos(0.5*np.sqrt(3) * x)

#grid of XY values
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)

# Calculation of eigenvalues of M for each (x,y) value
M_eig_1 = np.zeros((100, 100))
M_eig_2 = np.zeros((100, 100))
for i in range(100):
    for j in range(100):
        M = np.array([[0, F(X[i, j], Y[i, j])],
                      [F_conj(X[i, j], Y[i, j]), 0]], dtype=complex)
        eigvals = np.linalg.eigvalsh(M)
        M_eig_1[i, j] = eigvals[0]
        M_eig_2[i, j] = eigvals[1]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize=(12,4))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, M_eig_1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax1.set_title('Eigenvalues_1')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, M_eig_2, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax2.set_title('Eigenvalues_2')

plt.show()
```
